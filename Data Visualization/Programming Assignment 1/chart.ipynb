```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
```

```python
# Data Loader
df = pd.read_csv('./ExcelFormattedGISTEMPData2CSV.csv')
df.head()
```

```python
sns.lineplot(x = df['Year'], y = df['Glob'], label='Global', color='purple')
sns.lineplot(x = df['Year'], y = df['NHem'], label='Northern Hemisphere')
sns.lineplot(x = df['Year'], y = df['SHem'], label='Sorthern Hemisphere')
plt.xlabel('Year')
plt.ylabel('Deviation')
plt.legend()
plt.title('The GISTEMP gragh for the Globe, the North and South Hemispheres')
plt.savefig('chart.png')
```
